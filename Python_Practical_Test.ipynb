# SECTION A: Core OOP Concepts (Classes, Objects)

```python
#Define a class Car with attributes brand, model, and year. Include a method get_description() that returns a formatted string describing the car.
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    
    def get_description(self):
        return f"My car is a {self.year} {self.brand} {self.model}"
```

```python
#Create an instance of the Car class for a "Toyota Corolla" manufactured in 2020.
my_car = Car("Toyota", "Corolla", 2020) 
#Print the description of the car using the get_description() method.
print(my_car.get_description())  # Output: 2020 Toyota Corolla
```

```python
#Create a Garage class with methods to manage cars.
class Garage:
    def __init__(self):
        self.cars = []  # List to hold Car objects
    
    def add_car(self, car):
        self.cars.append(car)  # Adds a car to the garage
    
    def get_cars_by_brand(self, brand):
        return [car for car in self.cars if car.brand == brand]
    
# Example usage:
garage = Garage()
garage.add_car(car1)
garage.add_car(car2)
garage.add_car(car3)

# Get all cars by Toyota
toyota_cars = garage.get_cars_by_brand("Toyota")
for car in toyota_cars:
    print(car.get_description())  
```

# SECTION B: Inheritance

```python
#Create a base class Animal with attributes name and species. Add a method make_sound() that prints a generic animal sound.
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species
    
    def make_sound(self):
        print("Some generic animal sound")

#Create a subclass Dog that inherits from Animal. Override the make_sound() method to print "Woof!".
# Subclass Dog
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, species="Dog") 
        self.breed = breed

    def make_sound(self):
        print("Woof!")

#Create a subclass Cat with an additional attribute color and a method purr() that prints "Purr...".
# Subclass Cat
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name, species="Cat")  # Set species to "Cat"
        self.color = color

    def purr(self):
        print("Purr...")
#Create a list of Animal objects (some Dogs, some Cats). Use a loop to call make_sound() for each.
animals = [
    Dog(name="Buddy", breed="Golden Retriever"),
    Cat(name="Whiskers", color="Tabby"),
    Dog(name="Max", breed="Beagle"),
    Cat(name="Mittens", color="Black")
]
for animal in animals:
    animal.make_sound()  
#Write a function count_species(animals) that returns a dictionary with species as keys and counts as values.
def count_species(animals):
    species_count = {}
    for animal in animals:
        if animal.species in species_count:
            species_count[animal.species] += 1
        else:
            species_count[animal.species] = 1
    return species_count
species_counts = count_species(animals)
print(species_counts)  # Output: {'Dog': 2, 'Cat': 2}
```

```python
#Create a class Recipe with a dictionary to store ingredient: quantity. Add methods to:

class Recipe:
    def __init__(self, name):
        self.name = name
        self.ingredients = {}  # Dictionary to store ingredient: quantity
        self.created_date = None  # Will store the created date as a tuple (day, month, year)
    
    def add_ingredient(self, name, qty):
        """Add ingredient with its quantity to the recipe."""
        self.ingredients[name] = qty
    
    def total_ingredients(self):
        """Returns the total number of unique ingredients in the recipe."""
        return len(self.ingredients)


#Create a class Cookbook with a list of Recipe objects. Add a method to list all ingredients across recipes.
class Cookbook:
    def __init__(self):
        self.recipes = []  # List to store Recipe objects
        self.recipe_names = set()  # Set to store unique recipe names
    
    def add_recipe(self, recipe):
        """Add a recipe to the cookbook."""
        if recipe.name not in self.recipe_names:
            self.recipes.append(recipe)
            self.recipe_names.add(recipe.name)
    
    def list_all_ingredients(self):
        """List all unique ingredients across all recipes."""
        all_ingredients = set()
        for recipe in self.recipes:
            all_ingredients.update(recipe.ingredients.keys())  # Add all ingredients from each recipe
        return all_ingredients
# In Cookbook, maintain a set of all unique recipe names.
    def get_recipe_names(self):
        """Return a list of unique recipe names."""
        return list(self.recipe_names)
#Add a tuple to Recipe to store the created_date in the form (day, month, year).
class Recipe:
    def __init__(self, name, created_date=None):
        self.name = name
        self.ingredients = {}  # Dictionary to store ingredient: quantity
        self.created_date = created_date  # Tuple to store the creation date (day, month, year)
    
    def add_ingredient(self, name, qty):
        """Add ingredient with its quantity to the recipe."""
        self.ingredients[name] = qty
    
    def total_ingredients(self):
        """Returns the total number of unique ingredients in the recipe."""
        return len(self.ingredients)
```
